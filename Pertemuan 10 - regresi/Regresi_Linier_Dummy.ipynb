## Dummy of regression linear with sklearn and statsmodels

```python
# lib manipulation dataset
import numpy as np
import pandas as pd

# lib visualization data
import matplotlib.pyplot as plt
import seaborn as sns

# lib regression linear
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
```

### 1. Data Acquisition

```python
# load dataset
dataset = pd.read_csv("../dataset/dummy_regresi.csv", parse_dates=["tahun"])
dataset = dataset.set_index("tahun")
dataset.info()
```

```python
# show metadata
print(np.round(dataset.describe(),2))
```

### Exploration Data Analysis

```python
# function scatter plot
def scatter(x, y, label, title):
  
  # create scatter plots
  fig, ax = plt.subplots(figsize = (8,4))
  ax.scatter(x, y, color='tab:blue', label=label, linewidth=2)

  # set labels
  ax.set_title(title, fontsize=14)
  ax.set_xlabel("", fontsize=12)
  ax.set_ylabel("", fontsize=12)
  ax.legend(loc='upper left')
  ax.grid(True)
  
  # return values
  plt.tight_layout()
  plt.show()
```

```python
scatter(
  x=dataset["x1"],
  y=dataset["y"],
  label="label",
  title="title"
)
```

```python
scatter(
  x=dataset["x2"],
  y=dataset["y"],
  label="label",
  title="title"
)
```

```python
# func lineplot for timeseries
def lineplot(x, y, label, title):

  # create frame
  fig, ax = plt.subplots(figsize = (8,4))

  # time series plot
  ax.plot(x, y, color="tab:blue", label=label, linewidth=2, marker='o')

  # set label-labels
  ax.set_title(title,fontsize=14)
  ax.set_xlabel("",fontsize=12)
  ax.set_ylabel("",fontsize=12)
  ax.legend(loc="best")
  ax.grid(True)

  # rotasi label x menjadi 45 derajat
  plt.setp(ax.xaxis.get_majorticklabels(), rotation=30)
  plt.setp(ax.xaxis.get_majorticklabels(), rotation=30)
  plt.tight_layout()

  # return values
  return plt.show()
```

```python
# resample 4 monthly
dataset = dataset.resample('4ME', closed='left').sum()
print(dataset)
```

```python
# visualization x1
lineplot(
  x=dataset.index,
  y=dataset["x1"],
  title="Visualization of X1",
  label="variable x1"
)
```

```python
# visualization x2
lineplot(
  x=dataset.index,
  y=dataset["x2"],
  title="Visualization of X2",
  label="variable x2"
)
```

```python
# visualization y
lineplot(
  x=dataset.index,
  y=dataset["y"],
  title="Visualization of Y",
  label="variable y"
)
```
