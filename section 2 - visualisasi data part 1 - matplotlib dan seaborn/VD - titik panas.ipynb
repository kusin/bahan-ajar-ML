```python
# lib manipulasi data
import pandas as pd
import numpy as np

# lib visualisasi data
import seaborn as sns
import matplotlib.pyplot as plt

# lib praproses data
from sklearn.preprocessing import MinMaxScaler
```

1. Kebakaran Hutan

```python
# load dataset
dataset = pd.read_excel("../dataset/luas-karhutla-provinsi.xlsx", sheet_name="hasil")

# lihat metadataset
print(dataset.info())
```

```python
# create frame
fig, ax = plt.subplots(figsize = (8,4))

# barplot
sns.barplot(dataset, x="tahun", y="luas", color="tab:blue")

# set label-labels
ax.set_title("",fontsize=14)
ax.set_xlabel("",fontsize=12)
ax.set_ylabel("",fontsize=12)
#ax.legend(loc="best")
ax.grid(True)

# return values
plt.show()
```

2. Titik Panas

```python
# load dataset
dataset = pd.read_csv("../dataset/fire_archive_M-C61_484133.csv", parse_dates=["acq_date"])

# lihat metadataset
print(dataset.info())
```

```python
# Menghitung jumlah titik panas berdasarkan tanggal
dataset = dataset.groupby(['acq_date']).size().reset_index(name='hotspot')
print(dataset)
```

```python
# create frame
fig, ax = plt.subplots(figsize = (12,4))

# time series plot
ax.plot(
  dataset["acq_date"], dataset["hotspot"], 
  color="tab:blue", label="Hotspot at 2001 - 2023", linewidth=2
)

# set label-labels
ax.set_title("Visualization of Hotspot Indonesia",fontsize=14)
ax.set_xlabel("",fontsize=12)
ax.set_ylabel("",fontsize=12)
ax.legend(loc="best")
ax.grid(True)

# return values
plt.show()
```
