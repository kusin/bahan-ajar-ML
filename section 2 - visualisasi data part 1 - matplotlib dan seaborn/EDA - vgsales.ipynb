### 2. Sum of video games sales from all regions

- TimeSeries Plot

```python
# proses grouping berdasarkan tahun dan global sales
df = dataset.groupby(by="Year", as_index=False)["Global_Sales"].sum()

# proses menghapus penjualan tahun 2017, 2020
df = df.drop(labels=[37,38], axis=0)
```

```python
# membuat frame
fig, ax = plt.subplots(figsize = (15,5))

# membuat time series plot
ax.plot(df["Year"], df["Global_Sales"], color="tab:blue", label="Global sales", linewidth=2)

# membuat label-label
ax.set_title("History of video games sales", fontsize=12)
ax.set_xlabel("", fontsize=10)
ax.set_ylabel("", fontsize=10)
ax.legend(loc='upper left')
ax.grid(True)
 
# menampilkan plot
plt.show()
```
