```python
# lib getdataset from yahoo
import yfinance as yf

# lib manipulation dataset
import pandas as pd
import numpy as np
```

```python
# Alternatif, 
# BTC-USD, ETH-USD, AMZN, AAPL, GOOG, MSFT
ticker      = yf.Ticker("BTC-USD")
startDate   = "2015-01-01"
endDate     = "2024-06-01"
```

```python
def getData(ticker, startDate, endDate):

  dataset = ticker.history(start=startDate, end=endDate).reset_index()

  dataset = dataset[["Date", "Open", "High", "Low", "Close"]]

  dataset["Date"] = pd.to_datetime(dataset["Date"], format="%d-%m-%Y")
  
  return dataset
```

```python
x = getData(ticker, startDate, endDate)
```

```python
x.info()
```

```python
# lib visualisasi data
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# create frame
fig, ax = plt.subplots(figsize = (8,4))

# time series plot
ax.plot(x["Date"], x["Open"], color="tab:green", label="Open Price", linewidth=2)

# set label-labels
#ax.set_title(title,fontsize=14)
ax.set_xlabel("",fontsize=12)
ax.set_ylabel("",fontsize=12)
ax.legend(loc="best")
ax.grid(True)
```
