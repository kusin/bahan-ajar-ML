```python
# Mengimpor library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Memuat dataset IRIS
iris = load_iris()
X = iris.data

# Membuat dan melatih model K-Means dengan 3 klaster
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot hasil klastering
plt.figure(figsize=(8, 6))

# Plot sepal length vs sepal width
plt.subplot(2, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', marker='o', edgecolor='k')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Plot sepal length vs petal length
plt.subplot(2, 2, 2)
plt.scatter(X[:, 0], X[:, 2], c=y_kmeans, cmap='viridis', marker='o', edgecolor='k')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

# Plot sepal width vs petal length
plt.subplot(2, 2, 3)
plt.scatter(X[:, 1], X[:, 2], c=y_kmeans, cmap='viridis', marker='o', edgecolor='k')
plt.title('Sepal Width vs Petal Length')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Length')

# Plot petal length vs petal width
plt.subplot(2, 2, 4)
plt.scatter(X[:, 2], X[:, 3], c=y_kmeans, cmap='viridis', marker='o', edgecolor='k')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')

plt.tight_layout()
plt.show()

# Menampilkan hasil klastering
for i in range(len(y_kmeans)):
    print(f"Data ke-{i+1}: Klaster={y_kmeans[i]}")
```

```python
# Mengimpor library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Memuat dataset IRIS
iris = load_iris()
X = iris.data

# Mencari jumlah klaster optimal dengan metode elbow
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

# Plot metode elbow
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Metode Elbow untuk Menentukan Jumlah Klaster Optimal')
plt.xlabel('Jumlah Klaster')
plt.ylabel('Sum of Squared Distances (SSE)')
plt.grid(True)
plt.show()

# Membuat dan melatih model K-Means dengan jumlah klaster optimal (misalnya, 3 klaster)
optimal_k = 3  # ini contoh saja, Anda bisa menentukan dari plot elbow
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot hasil klastering
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', marker='o', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.title(f'Klastering IRIS menggunakan K-Means dengan {optimal_k} klaster')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid(True)
plt.show()

# Menampilkan hasil klastering
for i in range(len(y_kmeans)):
    print(f"Data ke-{i+1}: Klaster={y_kmeans[i]}")
```

```python
import warnings
warnings.filterwarnings("ignore")

# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Memuat dataset Iris
iris = load_iris()
X = iris.data

# Menentukan range untuk jumlah klaster
range_n_clusters = range(1, 11)
inertia = []

# Menggunakan metode Elbow untuk menemukan jumlah klaster optimal
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot hasil Elbow
plt.figure(figsize=(8, 6))
plt.plot(range_n_clusters, inertia, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()

# Menentukan jumlah klaster optimal dari hasil Elbow
optimal_clusters = 3

# Menerapkan KMeans dengan jumlah klaster optimal
kmeans = KMeans(n_clusters=optimal_clusters)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Visualisasi hasil klastering
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', marker='o')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='x')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KMeans Clustering of Iris Dataset')
plt.show()
```
