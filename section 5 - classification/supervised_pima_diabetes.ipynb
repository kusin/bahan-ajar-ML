```python
import pandas as pd
import matplotlib.pyplot as plt

# Mengimpor data dari file CSV (misalnya 'pima_diabetes.csv')
# data = pd.read_csv('pima_diabetes.csv')
data = dataset

# Membuat fungsi untuk menghitung rata-rata glukosa berdasarkan rentang umur
def calculate_glucose_mean_by_age_range(df, age_start, age_end):
    filtered_data = df[(df['Age'] >= age_start) & (df['Age'] <= age_end)]
    glucose_mean = filtered_data['Glucose'].mean()
    return glucose_mean

# Rentang umur yang ingin Anda tentukan
age_ranges = [(21, 30), (31, 40), (41, 50), (51, 60), (61, 70), (71, 80)]

# Menghitung rata-rata glukosa untuk setiap rentang umur
glucose_means = [calculate_glucose_mean_by_age_range(data, start, end) for start, end in age_ranges]

# Membuat plot
plt.figure(figsize=(10, 6))
plt.bar(range(len(age_ranges)), glucose_means, color='skyblue')
plt.xlabel('Rentang Umur')
plt.ylabel('Rata-rata Glukosa')
plt.title('Rata-rata Glukosa Berdasarkan Rentang Umur')
plt.xticks(range(len(age_ranges)), [f'{start}-{end}' for start, end in age_ranges])
plt.show()
```
